## Q2. Hydrostatic Balance

Starting from the vertical momentum equation for a resting atmosphere, derive the hydrostatic balance relation:

\[ \frac{dp}{dz} = - \rho g \]

### Tasks:
1. State the physical assumptions (1-2 lines).  
2. Show the key steps leading to the above equation (3-6 lines).  
3. In 3-4 sentences, interpret the physical meaning of hydrostatic balance.

*(Write your derivation and explanation here.)*

The hydrostatic balance relation, $\frac{dp}{dz} = - \rho g$, is derived from the vertical momentum equation for a fluid by assuming that the atmosphere is effectively at rest (i.e., neglecting vertical acceleration, $Dw/Dt \approx 0$) and ignoring viscous forces. This simplification reduces the full momentum equation to a simple balance where the upward pressure gradient force ($-\frac{dp}{dz}$) is precisely opposed by the downward force of gravity or weight per unit volume ($\rho g$). Physically, this equation is a statement of vertical equilibrium and is crucial for explaining why atmospheric pressure decreases with altitude, as the pressure at any level must be great enough to support the total weight of the column of air above it.

```python
# Q3. Pressure-Height Relationship in an Isothermal Atmosphere

# Temperature constant
T = 288.0   # Kelvin

# Scale height
H = R * T / g
print(f"Scale height H = {H:.2f} m (~{H/1000:.2f} km)")

# Height range: 0-20 km
z = np.linspace(0, 20000, 201)

# Exponential pressure profile
p = p0 * np.exp(-z / H)

# Plot
plt.figure(figsize=(5,7))
plt.plot(p/100, z/1000)  # convert to hPa and km
plt.gca().invert_yaxis()
plt.xlabel("Pressure (hPa)")
plt.ylabel("Height (km)")
plt.title("Pressure Profile in an Isothermal Atmosphere")
plt.grid(True)
plt.show()
```

Pressure decreases exponentially with height because in an isothermal atmosphere density is proportional to pressure. This makes the rate of pressure decrease proportional to the pressure itself, which mathematically leads to an exponential profile.

The scale height is the characteristic vertical distance over which pressure (or density) drops by a factor of e. Physically, it measures how vertically extended the atmosphere is, depending on temperature, molecular mass, and gravity.

### (c) Stability and Aerosols  
Explain in 3-4 sentences why atmospheric stability is important for:
- Cloud formation  
- Vertical mixing of aerosols  
- Pollution events near the surface  

*(Write your answers here.)*

(a) Absolutely stable atmosphere: The environmental lapse rate (ELR) is less than the dry adiabatic lapse rate (~9.8 K/km), so temperature decreases slowly with height.

Neutral atmosphere: The ELR is equal to the dry adiabatic lapse rate, so a displaced parcel has the same temperature as its surroundings.

Absolutely unstable atmosphere: The ELR is greater than the dry adiabatic lapse rate, so temperature decreases very rapidly with height.

(b) Absolutely stable: A lifted parcel becomes colder and denser than the environment and sinks back to its original level.

Neutral: A lifted parcel remains at its new height because it has no buoyancy relative to its surroundings.

Absolutely unstable: A lifted parcel stays warmer and less dense than the environment and continues to rise.

(c) Atmospheric stability strongly controls vertical motion in the atmosphere. In unstable conditions, rising air promotes cloud formation and efficiently mixes aerosols upward, dispersing pollutants. Stable conditions suppress vertical motion, limiting cloud development and trapping aerosols near the surface. This can lead to high pollution concentrations and poor air quality, especially during temperature inversions.

## Q5. Link Between Atmospheric Dynamics and Aerosols

In 4‚Äì6 sentences, discuss how atmospheric structure and stability influence:

- Vertical distribution of aerosols  
- Near-surface pollution events (e.g., winter smog)  
- Interpretation of AOD measurements from satellites  

Use a simple example such as:
- Stable winter morning boundary layer  
- Convective afternoon boundary layer

*(Write your answer here.)*

Atmospheric structure and stability control how efficiently aerosols are mixed vertically. In a stable winter morning boundary layer, weak turbulence and temperature inversions suppress vertical motion, trapping aerosols near the surface and leading to high near-surface pollution such as winter smog. In contrast, a convective afternoon boundary layer is unstable, with strong turbulent mixing that redistributes aerosols through a deeper layer and reduces surface concentrations. This vertical redistribution affects satellite-retrieved aerosol optical depth (AOD), since AOD measures the column-integrated aerosol amount rather than surface pollution alone. As a result, high AOD does not always imply poor surface air quality, especially under well-mixed conditions, while severe pollution can occur with modest AOD when aerosols are confined near the ground
