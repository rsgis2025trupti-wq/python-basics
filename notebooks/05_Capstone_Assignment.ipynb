# **Section 2 — Lists, Loops, and Conditions (Notebook 02)**

### **Q2.1 — Day-wise SST Printing**  
Using the `sst` list from Q1.2, print:

```
Day 1: value
Day 2: value
...
```


````python
# Q2.1
sst = [27.9,28.0,28.4,28.6,28.8,29.1,28.7,28.3,27.8] # sst list here

for i in range(len(sst)):
    print(f"Day {i+1}: {sst[i]}")
````

### **Q2.2 — Classification**  
Classify each SST into:
- **HOT** (> 29)  
- **WARM** (28–29)  
- **NORMAL** (< 28)

Print classification for each day.  
At the end, print counts for each category.


````python
# Q2.2
hot_count = 0
warm_count = 0
normal_count = 0


for i, temp in enumerate(sst):
    if temp > 29:
        print(f"Day {i+1}: HOT")
        hot_count += 1
    elif 28 <= temp <= 29:  # Corrected condition for WARM (inclusive)
        print(f"Day {i+1}: WARM")
        warm_count += 1
    else: # All other cases are NORMAL (temp < 28)
        print(f"Day {i+1}: NORMAL")
        normal_count += 1

print(f"\nCounts:")
print(f"HOT: {hot_count}")
print(f"WARM: {warm_count}")
print(f"NORMAL: {normal_count}")
````

### **Q2.3 — Rolling 3-day Mean**  
Compute a 3-day rolling mean using loops (without NumPy).  
Example output:
```
Mean Day1-3: ...
Mean Day2-4: ...
...
```


````python
# Q2.3
sst = [27.9,28.0,28.4,28.6,28.8,29.1,28.7,28.3,27.8]

for i in range(len(sst) - 2):
  current_window_sum = sst[i] + sst[i+1] + sst[i+2]
  rolling_mean = current_window_sum / 3
  print(f"Mean Day{i+1}-{i+3}: {rolling_mean:.2f}")
````

### **Q3.3 — Convert to NumPy + Plot**  

1. Convert `sst` to a NumPy array  
2. Compute anomalies  
3. Create a line plot of SST  
4. Create a second plot of anomalies  
5. Include:
   - Titles  
   - Axis labels  
   - Grid  
   - A horizontal zero line for anomalies  


````python
# Q3.3
import numpy as np
import matplotlib.pyplot as plt

sst = np.array([28.1, 28.3, 28.4, 28.0, 27.9, 28.5, 28.7, 28.6, 28.2, 28.1])
mean_sst = np.mean(sst)
sst_anom = sst - mean_sst # Directly compute anomalies, no need for np.array again if sst is already np array

days = np.arange(1, len(sst) + 1)

fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Plot 1: SST
ax[0].plot(days, sst, marker='o')
ax[0].set_title("Daily SST")
ax[0].set_ylabel("SST (°C)")
ax[0].grid(True)

# Plot 2: Anomalies
ax[1].plot(days, sst_anom, marker='o', color='red')
ax[1].axhline(0, color='black', linestyle='--')
ax[1].set_title("Daily SST Anomalies")
ax[1].set_xlabel("Day")
ax[1].set_ylabel("SST anomaly (°C)")
ax[1].grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()

````

# **Section 4 — NetCDF + xarray: Remote Sensing Dataset (Notebook 04)**

A small SST NetCDF file will be downloaded automatically.



````python
# Auto-download NetCDF file
import requests, os

url = "https://raw.githubusercontent.com/EarthSystem-Science-Lab/python-basics/main/data/data_sst.nc"
local = "data_sst.nc"

if not os.path.exists(local):
    r = requests.get(url)
    open(local,"wb").write(r.content)

ds = xr.open_dataset(local)
ds
````

### **Q4.3 — Time-Mean Map**  
1. Compute `sst_mean_time = ds.sst.mean(dim='time')`  
2. Plot a colour map (lat vs lon)  
3. Add labels, title, and colorbar  



````python
# Q4.3
sst_mean_time = ds.sst.mean(dim='time')
sst_mean_time.plot()

plt.figure()
plt.plot("sst_mean_time")
plt.axhline(0, color='black', linestyle='')
plt.xlabel("lon")
plt.ylabel("lat")
plt.title("Time-Mean Sea Surface Temperature")
plt.show()
````

### **Q4.4 — Time Series at a Point**  
Choose a latitude & longitude inside file domain.  
Extract nearest time series and plot it.

Add axis labels and title.


````python
# Q4.4
latitude = 16
longitude = 10

plt.figure
plt.plot(ds.sst.sel(lat=latitude, lon=longitude, method='nearest'))
plt.xlabel("Time")
plt.ylabel  ("SST (°C)")
plt.title(f"Time Series at {latitude}°N, {longitude}°E")
plt.show()


````
