### **Q1.2 — Basic Statistics**  
Given:
```python
sst = [27.9, 28.0, 28.4, 28.6, 28.8, 29.1, 28.7, 28.3, 27.8]
```

Compute:
- Mean  
- Median  
- Minimum  
- Maximum  
- Standard deviation  

Print with clear labels.


````python
# Q1.2
sst = [27.9,28.0,28.4,28.6,28.8,29.1,28.7,28.3,27.8]

mean_sst = np.mean(sst)
median_sst = np.median(sst)
min_sst = np.min(sst)
max_sst = np.max(sst)
std_sst = np.std(sst)

print(f"Mean: {mean_sst}")
print(f"Median: {median_sst}")
print(f"Minimum: {min_sst}")
print(f"Maximum: {max_sst}")
print(f"Standard Deviation: {std_sst}")
````

### **Q2.2 — Classification**  
Classify each SST into:
- **HOT** (> 29)  
- **WARM** (28–29)  
- **NORMAL** (< 28)

Print classification for each day.  
At the end, print counts for each category.


````python
# Q2.2
hot_count = 0
warm_count = 0
normal_count = 0


for i, temp in enumerate(sst):
    if temp > 29:
        print(f"Day {i+1}: HOT")
        hot_count += 1
    elif 28 <= temp <= 29:  # Corrected condition for WARM (inclusive)
        print(f"Day {i+1}: WARM")
        warm_count += 1
    else: # All other cases are NORMAL (temp < 28)
        print(f"Day {i+1}: NORMAL")
        normal_count += 1

print(f"\nCounts:")
print(f"HOT: {hot_count}")
print(f"WARM: {warm_count}")
print(f"NORMAL: {normal_count}")
````
