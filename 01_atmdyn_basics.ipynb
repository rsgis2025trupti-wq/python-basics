## Q2. Hydrostatic Balance

Starting from the vertical momentum equation for a resting atmosphere, derive the hydrostatic balance relation:

\[ \frac{dp}{dz} = - \rho g \]

### Tasks:
1. State the physical assumptions (1-2 lines).  
2. Show the key steps leading to the above equation (3-6 lines).  
3. In 3-4 sentences, interpret the physical meaning of hydrostatic balance.

*(Write your derivation and explanation here.)*

The hydrostatic balance relation, $\frac{dp}{dz} = - \rho g$, is derived from the vertical momentum equation for a fluid by assuming that the atmosphere is effectively at rest (i.e., neglecting vertical acceleration, $Dw/Dt \approx 0$) and ignoring viscous forces. This simplification reduces the full momentum equation to a simple balance where the upward pressure gradient force ($-\frac{dp}{dz}$) is precisely opposed by the downward force of gravity or weight per unit volume ($\rho g$). Physically, this equation is a statement of vertical equilibrium and is crucial for explaining why atmospheric pressure decreases with altitude, as the pressure at any level must be great enough to support the total weight of the column of air above it.

```python
# Q3. Pressure-Height Relationship in an Isothermal Atmosphere

# Temperature constant
T = 288.0   # Kelvin

# Scale height
H = R * T / g
print(f"Scale height H = {H:.2f} m (~{H/1000:.2f} km)")

# Height range: 0-20 km
z = np.linspace(0, 20000, 201)

# Exponential pressure profile
p = p0 * np.exp(-z / H)

# Plot
plt.figure(figsize=(5,7))
plt.plot(p/100, z/1000)  # convert to hPa and km
plt.gca().invert_yaxis()
plt.xlabel("Pressure (hPa)")
plt.ylabel("Height (km)")
plt.title("Pressure Profile in an Isothermal Atmosphere")
plt.grid(True)
plt.show()
```
